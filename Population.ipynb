```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_76034.csv', skiprows=4)
```

```python
#Checking the columns and previewing the data
data.head()
data.columns
```

```python
# Extracting country names and their 2022 population
population_2023 = data[['Country Name', '2023']]

#Remove NaN or missing values:
population_2023 = population_2023.dropna()
```

```python
#Sort and Select Top 10 Most Populated Countrie
top_10 = population_2023.sort_values(by='2023', ascending=False).head(10)
```

```python
#Plot the bar chart---
#Using Matplotlib
plt.figure(figsize=(12,6))
plt.bar(top_10['Country Name'], top_10['2023'], color='teal')
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.title('Top 10 Most Populated Countries in 2023')
plt.tight_layout()
plt.show()

#Using Seaborn
plt.figure(figsize=(12,6))
sns.barplot(x='Country Name', y='2023', data=top_10)
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.title('Top 10 Most Populated Countries in 2023')
plt.tight_layout()
plt.show()
```
